```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = Image.open("Data/linux.jpg")
```

```python
plt.figure(figsize=(6,6))
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
gray_img = img.convert('L')

plt.figure(figsize=(6,6))
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
width, height = img.size
print(f"Original image size: width={width}, height={height}")
```

```python
gray_img = gray_img.resize((128, 128))
```

```python
img_array = np.array(gray_img)

print("image shape: {}" .format(img_array.shape))
print("Output matrix:")
print(img_array)
```

```python
np.savetxt("Data/Lenna.txt", img_array, fmt='%d')
```
